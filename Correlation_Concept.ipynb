**Correlation** :
measures the strength and direction of a linear relationship between two variables.



image.png

Strength of r from -1 to 1.

Value close to 1 implies a strong linear relationship and vice versa, and around 0 implies no linear relationship.

Use Case:

Determining relationship between consumer income and expenditure on luxury items.

In finance, correlating different stock movements to diversify the portfolio.





**Correlation Matrix**

Quantifies the degree of correlation coefficients(Pearson, Spearmans, KendallsTau) between set of variables.

Real life Use Cases:

Finance: Correlation matrices are used to understand the relationships between different stocks or financial instruments, aiding in portfolio diversification.

Marketing: Analyzing customer data to find correlations between different buying behaviors.

Healthcare: Studying correlations between different lifestyle factors and health outcomes.

```python
import numpy as np
import pandas as pd
```

```python
# np.random.seed(42)  same random numbers

data={
    'Age' : np.random.normal(30,10,100).astype(int),
    'Annual Income (K$)':  np.random.normal(50,20,100).astype(int),
    'Spending Score (1-100)' : np.random.randint(1,100,100),
    'Years with Company' :np.random.normal(5,2,100).astype(int)
}

df=pd.DataFrame(data)
df
```

```python
from tabulate import tabulate
```

```python
correlation_matrix=df.corr()

# print(tabulate(correlation_matrix,headers='keys'))
#print correlation matrix as table
print(tabulate(correlation_matrix, headers='keys',tablefmt='grid',numalign='right',floatfmt='.2f'))
```
